### Задача 1

Пусть $X_1,...,X_n$ выборка из из распределения $U(0, \theta)$. На лекции во время практики вы уже визуализировали свойство несмещенности для следующих оценок параметра $\theta$: $2\overline{X}$, $X_{(n)}$, $\frac{n+1}{n}X_{(n)}$. Теперь наша задача убедиться, согласуется ли практика с теорией. Для решения этой задачи воспользуйтесь утверждениями с лекции.

*Указание.* Найдите функцию распределения случайной величины $X_{(n)}$, посчитав вероятность того, что максимальное значение среди элементов выборки не превосходит числа $x$. Используйте при этом определение выборки. После чего получите плотность $X_{(n)}$.


$X_{(n)} = max \{X_i, i=1,...,n\}$. Тогда функция распределения случайной величины $X_{(n)}$ есть
\begin{equation*}
F(x) = P(X_{(n)} \leq x) = P(X_i \leq x, \forall i = 1,...,n).
\end{equation*}
Так как случайные величины выборки независимые и одинаково распределённые, то
\begin{equation*}
P(X_i \leq x, \forall i = 1,...,n) = \displaystyle\prod_{i=1}^{n} P(X_i \leq x, i = 1,...,n).
\end{equation*}
Для равномерного распределения $U(0, \theta)$ $\forall i=1,...n$ справедливо, что $P(X_i \leq x, i = 1,...,n) = \frac{x}{\theta}$. Тогда
\begin{equation*}
F(x) = P(X_{(n)} \leq x) =  \displaystyle\prod_{i=1}^{n} P(X_i \leq x, i = 1,...,n) = \Big(\frac{x}{\theta}\Big)^n.
\end{equation*}
Тогда плотность распределения случайной величины $X_{(n)}$ есть
\begin{equation*}
f_{X_{(n)}}(x) = \frac{d}{dx}F(x) = n \frac{x^{n-1}}{\theta^n}.
\end{equation*}.

Теперь найдём математическое ожидание случайной величины $X_{(n)}$
\begin{equation*}
\mathbb{E}[X_{(n)}] =  \int_{0}^{\theta} x n \frac{x^{n-1}}{\theta^n}  \,dx = \frac{n}{\theta^n}\int_{0}^{\theta} x^n\,dx = \frac{n\theta}{n+1}.
\end{equation*}
Отсюда видим, что оценка $X_{(n)}$ не является несмещённой, а оценка $\frac{n+1}{n}X_{(n)}$ - несмещённая, так как
\begin{equation*}
\mathbb{E}\Big[\frac{n+1}{n}X_{(n)}\Big] = \frac{n+1}{n}\mathbb{E}[X_{(n)}] = \frac{n+1}{n} \cdot  \frac{n\theta}{n+1} = \theta.
\end{equation*}
Посмотрим, является ли несмещённой оценка $2\overline{X}$. Для равномерного распределения
\begin{equation*}
\mathbb{E}[X_i] = \frac{\theta}{2}, \forall i = 1,...,n.
\end{equation*}
Тогда 
\begin{equation*}
\mathbb{E}[2\overline{X}] = 2\mathbb{E}[\overline{X}] = 2\mathbb{E}\Big[\frac{1}{n}\displaystyle\sum_{i=1}^{n}X_i\Big] = \frac{2}{n}\displaystyle\sum_{i=1}^{n}\mathbb{E}[X_i] = \frac{2}{n} \cdot n \cdot \mathbb{E}[X_i] = 2 \cdot \frac{\theta}{2} = \theta.
\end{equation*}

Значит,  оценка $2\overline{X}$ является несмещённой.


### Задача 2


Пусть $X_1,...,X_n$ выборка из некоторого распределения $\mathrm{P}$, причем ${\sf D} X_1=\sigma^2<+\infty$, и $\sigma$ неизвестно.
Рассмотрим статистику $S^2=\frac1n\sum\limits_{i=1}^n(X_i-\overline{X})^2$.

1. Докажите, что статистика $S^2$ равна $\overline{X^2}-\overline{X}^2$.


2. Является ли она состоятельной оценкой $\sigma^2$? Для ответа на этот вопрос воспользуйтесь утверждениями с лекции.


3. Является ли статистика $S^2$ несмещенной оценкой $\sigma^2$? Для ответа на этот вопрос вспомните различные варианты определения дисперсии.

1. Покажем, что $S^2 = \overline{X^2} - \overline{X}^2$.

$n \cdot S^2 = \displaystyle\sum_{i=1}^{n}(X_i - \overline{X})^2 = \displaystyle\sum_{i=1}^{n}X_i^2 - 2\overline{X}\displaystyle\sum_{i=1}^{n}X_i + n\overline{X}^2 = \displaystyle\sum_{i=1}^{n}X_i^2 - 2n\overline{X}^2 + n\overline{X}^2 = \displaystyle\sum_{i=1}^{n}X_i^2 - n\overline{X}^2$.
Тогда
\begin{equation*}
S^2 = \frac{1}{n} \displaystyle\sum_{i=1}^{n}X_i^2 - \overline{X}^2 = \overline{X^2} - \overline{X}^2.
\end{equation*}

2. Проверим, является ли статистика $S^2$ состоятельной оценкой дисперсии $\sigma^2$.
$S^2$ - состоятельная оценка $\sigma^2$, если $S^2$ сходится по вероятности к $\sigma^2$. По закону больших чисел
\begin{equation*}
\overline{X} \xrightarrow{P} \mathbb{E}[X_1],
\end{equation*}

\begin{equation*}
\overline{X^2} \xrightarrow{P} \mathbb{E}[X_1^2].
\end{equation*}

Тогда 
\begin{equation*}
S^2 = \overline{X^2} - \overline{X}^2 \xrightarrow{P} \mathbb{E}[X_1^2] - \Big(\mathbb{E}[X_1]\Big)^2 = \sigma^2,
\end{equation*}
следовательно,  статистика $S^2$ является состоятельной оценкой дисперсии $\sigma^2$.

3. Найдём математическое ожидание статистики $S^2$:

\begin{equation*}
\mathbb{E}[S^2] = \mathbb{E}\Big[\frac{1}{n}\displaystyle\sum_{i=1}^{n}(X_i - \overline{X})^2\Big] = \frac{1}{n}\mathbb{E}\Big[\displaystyle\sum_{i=1}^{n}(X_i - \overline{X})^2\Big].
\end{equation*}
Считаем известным, что $\mathbb{E}\Big[\displaystyle\sum_{i=1}^{n}(X_i - \overline{X})^2\Big] = (n-1)\sigma^2$.

Тогда 
\begin{equation*}
\mathbb{E}[S^2] = \frac{1}{n}\mathbb{E}\Big[\displaystyle\sum_{i=1}^{n}(X_i - \overline{X})^2\Big] = \frac{n-1}{n}\sigma^2 \neq \sigma^2,
\end{equation*}
значит, так как $\mathbb{E}[S^2] \neq \sigma^2$, $S^2$ не является несмещённой оценкой параметра $\sigma^2$.

### Задача 3


Пусть $X_1, ..., X_n$ &mdash; выборка из распределения $\mathcal{N}(0, \sigma^2)$. Известно, что в качестве оценки параметра $\sigma^2$ можно использовать следующие оценки $S^2, \frac{n}{n-1}S^2$. Проверьте эти оценки на несмещенность, повторив эксперимент с семинара.

Реализуйте две функции, каждая из которых на вход берет несколько выборок, а на выход выдает массив оценок (первая функция - $S^2$, вторая функция - $\frac{n}{n-1}S^2$)  для каждой выборки.  
*Примечание:* У матрицы, заданной в виде `numpy.array` есть метод `var` вычисляющий выборочную дисперсию.

```python
def estimate_s2(X):
    '''
    Принимает на вход массив размером (n_samples, sample_size), выдает массив оценок размера (n_samples,)
    '''
    s2_list = [np.var(X[i]) for i in range(len(X))]
    return s2_list

def estimate_s2_corrected(X):
    '''
    Принимает на вход массив размером (n_samples, sample_size), выдает массив оценок размера (n_samples,)
    '''
    s2_corrected_list = [np.var(X[i])*len(X[i])/(len(X[i])-1) for i in range(len(X))]
    return s2_corrected_list
```

Проверьте правильность своей реализации

```python
def est_plot(distribution, estimators, sample_size, sample_count):
    '''
    Построение графика разброса реализаций оценок и их средних значений.

    distribution -- распределение формата scipy.stats
    estimators -- список оценок и параметров для отрисовки графиков
    sample_size -- размер выборок
    sample_count -- количество генерируемых выборок
    '''
    X = distribution.rvs(size=(sample_count, sample_size))
    plt.figure(figsize=(15, 0.7*len(estimators)))
    for estimator, name, color, y in estimators:
        E = estimator(X)
        plt.scatter(E, np.zeros(sample_count) + y, alpha=0.1,
                        s=100, color=color, label=name)
        plt.scatter(np.mean(E), y, marker='*', s=300,
                        color='w', edgecolors='black')

        plt.vlines(1, -1, 1, color='r')
        plt.title('Размер выборки = %d' % sample_size)
        plt.yticks([])
        plt.legend()
    
```

```python
def mean_plot(distribution, estimators, n_grid, sample_count):
    '''
    distribution -- распределение формата scipy.stats
    estimators -- список оценок и параметров для отрисовки графиков
    n_grid -- массив значений размера выборки
    sample_count -- количество генерируемых выборок
    '''
    n_grid = n_grid.astype(int)
    X = distribution.rvs(size = (sample_count, len(n_grid)))
    plt.figure(figsize=(15, 3*len(estimators)))
    plt.hlines(1, 0, len(n_grid), color='r', label=r'$\theta$')
    for estimator, name, color, y in estimators:
        E = np.array([np.mean(estimator(X[0:len(n_grid), 0:(n_grid[i]-1)])) for i in range(len(n_grid))])
        plt.plot(n_grid, E, color= color, label= name)
        plt.xlabel('размер выборки')
        
        plt.legend()
        plt.title('зависимость среднего оценок от размера выборки')
        plt.xlim((10, len(n_grid)))
```

В данной задаче мы предполагаем, что распределение имеет экспоненциальный характер, оценим параметр этого распределения.

Для этого выведем ОМП для параметра экспоненциального распределения:

$\mathcal{L}(\lambda) = \displaystyle\prod_{i=1}^{n}f(x_i|\lambda)$, где $f(x|\lambda) = \lambda e^{-\lambda x}$ - плотность экспоненциального распределения. Тогда
\begin{equation*}
\mathcal{L}(\lambda) = \displaystyle\prod_{i=1}^{n}f(x_i|\lambda) = \lambda^n \cdot exp \Big(-\lambda \displaystyle\sum_{i=1}^{n} x_i\Big)
\end{equation*}
Найдём логарифм функции правдоподобия 
\begin{equation*}
\ln{\mathcal{L}(\lambda)} = n \ln \lambda - \lambda  \displaystyle\sum_{i=1}^{n} x_i
\end{equation*}
Теперь найдём оценку $\hat{\lambda}$, приравняв производную логарифма к нулю:
\begin{equation*}
\frac{d}{d\lambda}\ln{\mathcal{L}(\lambda)} = \frac{n}{\lambda} - \displaystyle\sum_{i=1}^{n} x_i = 0,
\end{equation*}

Откуда 
\begin{equation*}
\hat{\lambda} = \frac{n}{\displaystyle\sum_{i=1}^{n} x_i} = \frac{1}{\overline{x}}.
\end{equation*}

  является несмещенной оценкой параметра $\theta$. Будет ли выполнено свойство несмещенности для Ridge-регрессии?

  *Указание:* используйте свойство дисперсии в матричном виде.


**Напоминание**

1. Матричное свойство математического ожидания:

  $∀ A \in \mathbb{R}^{m \times k}$, случайного вектора $\psi \in \mathbb{R}^k$

  $$E A \psi = A E \psi$$
2. Матричное свойство дисперсии:
  $∀ A \in \mathbb{R}^{m \times k}$, случайного вектора $\psi \in \mathbb{R}^k$

  $$D A \psi = A (D \psi) A^T$$

1.Cлучайной величиной является шум $\varepsilon$, а также зависимый от случайной величины $\varepsilon$ вектор откликов $Y$. Матрица объект-признак $X$ - известна, а вектор $\theta$ - неизвестный вектор параметров.

2. Пусть $\hat{\theta} = (X^TX)^{-1} X^T Y$. Подставим $Y = X\theta + \varepsilon$. Тогда 
\begin{equation*}
\hat{\theta} = (X^TX)^{-1} X^T Y = (X^TX)^{-1} X^T \Big(X\theta + \varepsilon\Big) = \theta + (X^TX)^{-1} X^T \varepsilon.
\end{equation*}
Можем найти математическое ожидание
\begin{equation*}
\mathbb{E}[\hat{\theta}] = \theta + (X^TX)^{-1} X^T \mathbb{E}[\varepsilon] = \theta,
\end{equation*}
так как $\mathbb{E}[\varepsilon] = 0$. Поэтому $\mathbb{E}[\hat{\theta}] = \theta$, значит оценка $\hat{\theta} = (X^TX)^{-1} X^T Y$ несмещённая.

Для Ridge-регрессии: 
\begin{equation*}
\hat{\theta}_{Ridge} = (X^TX + \lambda I)^{-1} X^T X \theta + (X^TX + \lambda I)^{-1} X^T \varepsilon.
\end{equation*}
Тогда математическое ожидание 
\begin{equation*}
\mathbb{E}[\hat{\theta}_{Ridge}] = (X^TX + \lambda I)^{-1} X^T X \theta \neq \theta,
\end{equation*}
значит оценка $\hat{\theta}_{Ridge}$ смещённая для Ridge-регрессии.

3. По свойству дисперсии
\begin{equation*}
\mathbb{D}[\hat{\theta}] = \mathbb{D}[(X^TX)^{-1} X^T \varepsilon] = (X^TX)^{-1}X^T \mathbb{D}[\varepsilon] X (X^T X)^{-1},
\end{equation*}

\begin{equation*}
\mathbb{D}[\varepsilon] = \sigma^2 I
\end{equation*}
Тогда
\begin{equation*}
\mathbb{D}[\hat{\theta}] = \sigma^2 (X^TX)^{-1}X^T X (X^T X)^{-1} = \sigma^2 (X^TX)^{-1}.
\end{equation*}

Если признаки мультиколлинеарны, матрица $X^T X$ близка к вырожденной, а элемента её обратной матрицы $(X^T X)^{-1}$ велики, что приводит к увеличению дисперсии оценки $\hat{\theta}$ и делает её неустойчивой.
