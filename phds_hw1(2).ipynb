В этой задаче вам предстоит визуализировать центральную предельную теорему.

*a).* Пусть $\xi_1, \xi_2,..., \xi_n$ — независимые случайные величины, имеющие распределение $Exp(\lambda)$. Согласно центральной предельной теореме, имеет место сходимость $$Z_n = \frac{X_n - \mathsf{E}X_n}{\sqrt{\mathsf{D}X_n}} \stackrel{d}{\longrightarrow} \mathcal{N}(0, 1),$$ где $X_n = \sum_{i=1}^n \xi_i$.

Чтобы убедиться в этом, вам необходимо сгенерировать множество наборов случайных величин и рассчитать значение $Z_n$ для каждого набора, в зависимости от его размера.

Сгенерируйте 1000 наборов случайных величин $\xi_1^j,..., \xi_{300}^j$ из распределения $Exp(1)$.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sps
import sklearn as skl
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
size = 300  # размер выборки
samples_count = 1000  # количество выборок

sample = np.array([np.random.exponential(1, size) for i in range(samples_count)])
```

По каждому из них посчитайте сумму $X_{jn} = \sum\limits_{i=1}^n \xi_i^j$ для $1 \leqslant n \leqslant 300$, то есть сумма первых $n$ величин $j$-го набора.

```python
sums = np.cumsum(sample, 1)
```

Для этого среднего посчитайте величину $Z_{jn} = \frac{X_{jn} - \mathsf{E}X_{jn}}{\sqrt{\mathsf{D}X_{jn}}}$.

*Подсказка: можно воспользоваться возможностями библиотеки `numpy`.*

```python
z = []
for j in range(samples_count):
    z_j = []
    for n in range(size):
        z_j_n = (sums[j][n] - (n+1))/(np.sqrt(n+1))
        z_j.append(z_j_n)
    z.append(z_j)
```

Для каждого $j$ нанесите на один график зависимость $Z_{jn}$ от $n$ с помощью `plt.plot`. Каждая кривая должна быть нарисована *одним цветом* с прозрачностью `alpha=0.05`. Сходятся ли значения $Z_{jn}$ к какой-либо константе?

```python
fig, ax = plt.subplots(figsize = (20,8))
plt.grid()
for j in range(samples_count):
    plt.plot(np.arange(1,301), z[j], color = 'tab:orange', alpha = 0.05)
plt.xlabel('n - размер выборки')
plt.ylabel('$Z_{jn}$')
plt.legend()
```

Видим, что значения сходятся к нулю


Для $n=300$ по набору случайных величин $Z_{1,300}, ..., Z_{1000,300}$ постройте гистограмму. Похожа ли она на плотность распределения $\mathcal{N}(0, 1)$, которую тоже постройте на том же графике? Не забудьте сделать легенду (подписи на графиках, см. семинар и туториалы).

```python
z_ = [x[-1] for x in z]
plt.hist(z_, bins=30, density=True, alpha = 0.5, label='гистограмма')
x = np.linspace(min(z_), max(z_), 1000)
pdf = sps.norm.pdf(x, 0, 1)
plt.plot(x, pdf, 'tab:orange', label=f'Нормальное распределение N(0, 1)')
plt.xlabel('$z_{300}$')
plt.ylabel('Плотность вероятности')
plt.legend()
```

Видим, что при увеличении числа столбцов гистограммы, вершины столбцов всё лучше ложатся на график плотности вероятности нормального распределения.

<font color="blue">**Профиль физика**</font>

Время когерентности кубита подчиняется показательному распределению с математическим ожиданием 100 микросекунд. Найти вероятность того, что среднее время когерентности выборки из 100 кубитов составит не менее 90 микросекунд.

<font color="green">**Профиль биология**</font>

Время жизни лейкоцита подчиняется показательному распределению с математическим ожиданием 200 часов. Найти вероятность того, что среднее время жизни выборки из 100 лейкоцитов окажется не менее 180 часов.

Пусть $\tau$ - время когерентности одного кубита, тогда $\tau$ ~ Exp($\lambda$), где $\lambda$ = $\frac{1}{\mathbb{E}(\tau)}$. 
У нас $\mathbb{E}(\tau)$ = 100 мкс, при этом времена когерентности различных кубитов считаем независимыми, одинаково распределёнными случайными величинами. Стандартное отклоненение для случайной величины, подчиняющейся экспоненциальному распределению с $\mathbb{E}(\tau)$ = 100 мкс есть 

\begin{equation*}
\sigma = \mathbb{E}(\tau) = 100 ~\ мкс
\end{equation*}

Тогда согласно Центральной предельной теореме имеет место сходимость по распределению
\begin{equation*}
Z_n = \sqrt{n}\frac{\tau_n-\mathbb{E}(\tau)}{\sigma} \longrightarrow N(0,1),
\end{equation*}
где $\tau_n$ - выборочное среднее первых $n$ случайных величин $\tau$ (времён когерентности первых $n$ кубитов). Будем считать, что размер выборки $n$ = 100 достаточно большой, чтобы было справедливо
\begin{equation*}
Z_{100} = \sqrt{100}\frac{\tau_{100}-\mathbb{E}(\tau)}{\sigma} \sim N(0,1),
\end{equation*}
То есть будем считать, что плотность $Z_n$ такая же, как функция плотности вероятности для нормального распределения N(0,1). Подставляя в качестве $\tau_{100}$ 90 мкс, получим $Z_{100} = -1$.

Таким образом, 
\begin{equation*}
P(\tau_{100} \geqslant 90) = P(Z_{100} \geqslant -1) = 1 - P(Z_{100} < -1),
\end{equation*}
где значение $P(Z_{100} < -1)$ есть интеграл от плотности нормального распределения (функции распределения для случайной величины $Z_{100} \sim N(0,1)$) c -1 на верхнем пределе. Из табличных данных находим 
\begin{equation*}
P(Z_{100} < -1) = 0.1587
\end{equation*}
Тогда искомая вероятность есть
\begin{equation*}
P(\tau_{100} \geqslant 90) = P(Z_{100} \geqslant -1) = 1 - P(Z_{100} < -1) = 1 - 0.1587 = 0.8413 
\end{equation*}

> Как это может повлиять на работу kNN, почему?

**Ваш ответ:** В методе ближайших соседей оценивается расстояние до ближайших соседей в пространстве признаков. Соответственно, если какие-то признаки имеют существенно другой порядок величины, чем другие, они вносят больший вклад в расстояние между соседями, в результате веса одних признаков оказываются выше, чем других. Тогда kNN классификатор может посчитать ближайшими соседями просто те объекты, у которых наименьшее расстояние по признаку, который характеризуется наибольшим числом и практически не будет учитывать различие в остальных признаках, как итог, будет давать неверные предсказания.


Для решения этой проблемы нужно привести признаки к единому масштабу. Существует много методов масштабирования рассмотрим работу одного из них &mdash; **минимаксная нормализация**

  
> **Минимакс** &mdash; линейное преобразование данных в диапазоне $[0, 1]$, где минимальное и максимальное масштабируемые значения соответствуют 0 и 1 соответственно.
> Реализуется оно по следующей формуле:
>
> $$\widetilde{x}_{ij} = \frac{x_{ij} - m_j}{M_j - m_j}$$
>
> где $x_{ij}$ &mdash; значение признака $j$ для объекта $i$, а $m_j$ и $M_j$ &mdash; минимальное и максимальное значения признака $j$ в обучающем наборе данных.  
> В модуле `sklearn.preprocessing` есть [класс](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.MinMaxScaler.html) `MinMaxScaler`, который реализует эту операцию. Давайте рассмотрим методы этого класса подробнее:
>
> * `fit(X)`: вычисляет минимальные и максимальные значения для каждого признака.
>
> * `transform(X)`: применяет масштабирование к данным.
>
> * `fit_transform(X)`: объединяет оба метода, сначала подгоняя модель, а затем трансформируя данные.
>
> * `inverse_transform(X)`: восстанавливает исходные данные из масштабированных.

### Задача 4

<font color="blue">**Профиль физика**</font>  

Рассмотрим два атома в возбужденном состоянии, каждый из которых может спонтанно испустить фотон и перейти в основное состояние. Время жизни возбужденного состояния каждого атома $X_1$ и $X_2$ подчиняется экспоненциальному распределению с параметром $\lambda$, что характерно для процессов спонтанного испускания.  

Обозначим:  
- $m = \min(X_1, X_2)$ — время испускания первого фотона, соответствующее времени жизни менее устойчивого атома.  
- $M = \max(X_1, X_2)$ — время испускания второго фотона, соответствующее времени жизни более устойчивого атома.  

Найдите математическое ожидание и дисперсию $m$ и $M$.  


<font color="green">**Профиль биология**</font>

Рассмотрим два молекулярных белка в клетке, каждый из которых подвержен деградации под действием протеаз. Время жизни каждого белка $X_1$ и $X_2$ является случайной величиной, которая подчиняется экспоненциальному распределению с параметром $\lambda$, что соответствует характерному механизму распада белков в биологических системах.  

Обозначим:  
- $m = \min(X_1, X_2)$ — время деградации первого белка, что соответствует скорости разрушения менее стабильного белка.  
- $M = \max(X_1, X_2)$ — время деградации второго белка, которое характеризует устойчивость более долговечного белка.  

Найдите математическое ожидание и дисперсию $m$ и $M$.  

$X_1 \sim Exp(\lambda)$, $X_2 \sim Exp(\lambda)$, $m = min(X_1, X_2)$ - время испускания первого фотона, $M = max(X_1, X_2)$ - время испускания второго фотона. Тогда для функций распределения $X_1$ и $X_2$ имеем
\begin{equation*}
F_{X_1}(x)=P(X_1 \leq x) = 1-e^{-\lambda x}
\end{equation*}

\begin{equation*}
F_{X_2}(x)=P(X_2 \leq x) = 1-e^{-\lambda x}
\end{equation*}
Найдём функцию распределения $m = min(X_1, X_2)$

\begin{equation*}
F_m(x) = P(m \leq x) = P(min(X_1, X_2) \leq x),
\end{equation*}
но $P(min(X_1, X_2) \leq x)$ - вероятность того, что хотя бы одна из величин $X_1$ или $X_2$ не превосходит $x$. Тогда
\begin{equation*}
P(min(X_1, X_2) \leq x) = 1 - P(X_1 > x ~\ и ~\ X_2 > x).
\end{equation*}
Из независимости $X_1$ и $X_2$ имеем
\begin{equation*}
P(X_1 > x ~\ и ~\ X_2 > x) = P(X_1 > x) \cdot P(X_2 > x) = (1-P(X_1 \leq x))\cdot(1-P(X_2 \leq x)) = e^{-2\lambda x}.
\end{equation*}
Тогда 
\begin{equation*}
F_m(x) = P(m \leq x) = 1 - e^{-2\lambda x},
\end{equation*}
то есть $m \sim Exp(2\lambda)$. Значит, $\mathbb{E}(m) = \frac{1}{2\lambda}$ и $\mathbb{Var}(m) = \frac{1}{4\lambda^2}$. 

Теперь найдём распределение $M = max(X_1, X_2)$

\begin{equation*}
F_M(x) = P(M \leq x) = P(max(X_1, X_2) \leq x),
\end{equation*}
но $P(max(X_1, X_2) \leq x)$ - вероятность того, cразу обе величины $X_1$ и $X_2$ не превосходят $x$, то есть
\begin{equation*}
P(max(X_1, X_2) \leq x) = P(X_1 \leq x ~\ и ~\ X_2 \leq x) = P(X_1 \leq x) \cdot P(X_2 \leq x)
\end{equation*}
так как $X_1$ и $X_2$ независимы. 

\begin{equation*}
F_M(x) = P(max(X_1, X_2) \leq x) = P(X_1 \leq x) \cdot P(X_2 \leq x) = (1-e^{-\lambda x})^2.
\end{equation*}
Тогда плотность вероятности для $M$ есть 
\begin{equation*}
f_M(x) = p(x) = \frac{d}{dx}F_M(x) = \frac{d}{dx}(1-e^{-\lambda x})^2 = 2 \lambda e^{-\lambda x} (1-e^{-\lambda x})
\end{equation*}
Тогда можем найти математическое ожидание по определению 
\begin{equation*}
\mathbb{E}(M) =  \int_{0}^{+ \infty} xp(x) \,dx = \int_{0}^{+ \infty} 2x \lambda e^{-\lambda x} (1-e^{-\lambda x}) \,dx = 2\lambda \bigg[\frac{1}{\lambda}\int_{0}^{+ \infty} x \cdot \lambda e^{-\lambda x}  \,dx - \frac{1}{2\lambda}\int_{0}^{+ \infty} x \cdot 2\lambda e^{-2\lambda x} \,dx \bigg] = 2\lambda \bigg[\frac{1}{\lambda}\mathbb{E}(Exp(\lambda)) - \frac{1}{2\lambda}\mathbb{E}(Exp(2\lambda))\bigg] = \frac{3}{2\lambda}.
\end{equation*}
Аналогично для дисперсии 
\begin{equation*}
\mathbb{Var}(M) = \int_{0}^{+ \infty} x^2 p(x) \,dx - (\mathbb{E}(M))^2 = \frac{7}{2\lambda^2} - \frac{9}{4\lambda^2} = \frac{5}{4\lambda^2}.
\end{equation*}
